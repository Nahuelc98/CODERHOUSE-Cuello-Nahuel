Creación de columnas para facilitar un análisis más profundo y detallado:

Para lograr un análisis preciso y de mayor profundidad, es fundamental contar con información detallada sobre el porcentaje de éxito de disparos y la cantidad de duelos aereos en cada partido. Estos indicadores permitirán evaluar el rendimiento de los equipos en distintos aspectos del juego y extraer conclusiones más sólidas a partir de los datos.

En particular, resultan de gran interés las columnas "Éxito_de_Disparos" y "Total_Duelos_Aéreos", ya que permiten evaluar la eficacia ofensiva y el rendimiento en los enfrentamientos individuales aéreos, respectivamente.

```python
# Creación de la columna: "Exito_de_Disparos"
# Corresponde al porcentaje de éxito en los disparos, calculada como la división entre disparos al arco y disparos totales, luego se saca el porsentaje del resultado.

df['Exitos_Disparos'] = np.where(
    df['Disparos_Totales'] > 0,
    (df['Disparos_al_Arco'] / df['Disparos_Totales']) * 100,
    0
)

df['Exitos_Disparos'] = df['Exitos_Disparos'].round(2)

# Chequeo de la columna nueva:
print(df[['Exitos_Disparos']])
```

Se unificarán tres columnas en una sola con el objetivo de simplificar la lectura y facilitar el análisis del rendimiento de los equipos.

Las columnas involucradas son: "Victoria", "Empate" y "Derrota", las cuales actualmente están en formato binario (0 y 1).
Cada fila presenta un único valor en 1, representando el resultado correspondiente.

A partir de esta estructura, se creará una nueva columna denominada "V-E-D", que almacenará los valores en formato texto: "Victoria", "Empate" o "Derrota", según corresponda en cada caso.

```python
# Se transformo las tres columnas binarias ("Victoria", "Empate", "Derrota") en una sola columna de tipo texto que indique directamente el resultado.

def clasificar_resultado(fila):
    if fila['Victoria'] == 1:
        return 'Victoria'
    elif fila['Empate'] == 1:
        return 'Empate'
    elif fila['Derrota'] == 1:
        return 'Derrota'
    else:
        return 'Sin dato'

df['V-E-D'] = df.apply(clasificar_resultado, axis=1)
```

## 1. ¿Influye jugar de local en las probabilidades de ganar?

-- Se vincularán las columnas "Local_Visitante" y "V-E-D" para calcular el porcentaje de victorias, empates y derrotas según la condición de juego, y compararlos mediante un gráfico.

```python
# Agrupar por 'Local_Visitante' y contar resultados:
tabla = df.groupby('Local_Visitante')['V-E-D'].value_counts().unstack().fillna(0)

# Orden de condición:
tabla = tabla[['Victoria', 'Empate', 'Derrota']]

# Calcular porcentaje fila por fila por condición:
local_pct = tabla.div(tabla.sum(axis=1), axis=0) * 100

# Colores personalizados:
colores = ['green', 'yellow', 'red']

# Gráfico de barras apiladas:
local_pct.plot(kind="bar", stacked=True, color=colores, figsize=(8, 5))

# Títulos y Etiquetas:
plt.title("Influencia de la localía en el resultado del partido", fontsize=17, weight='bold')
plt.ylabel("Porcentaje", fontsize=14)
plt.xlabel("Condición", fontsize=14)

# Leyenda:
plt.legend(title="Resultado", title_fontsize=12, fontsize=10)

# Ejes:
plt.xticks(rotation=0, fontsize=14)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

-- Conclusiones:

1. Mayor probabilidad de victoria jugando de local.
  * El segmento verde (Victoria) es más alto en la barra de "Local" que en la   de "Visitante".
  * Esto sugiere que los equipos tienen más chances de ganar cuando juegan de local.
2. Más derrotas cuando se juega como visitante.
  * El segmento rojo (Derrota) es más grande en la barra de "Visitante".
  * Se interpreta que los equipos pierden más partidos jugando fuera de casa.
3. Empates relativamente similares.
  * El segmento amarillo (Empate) tiene proporciones similares en ambas condiciones, aunque parece ligeramente mayor en local.
  * Esto indica que la condición de local o visitante no afecta demasiado la probabilidad de empate.

-- Pregunta:

  * ¿Influye jugar de local en las probabilidades de ganar?

   * Si, la localía tiene un efecto positivo en el rendimiento de los equipos, aumentando las probabilidades de victoria y reduciendo las de derrota. Esto respalda la conocida hipótesis de que jugar de local ofrece una ventaja competitiva.

**De esta manera el DataFrame principal estará enriquecido con información de rendimiento de los equipos, lista para análisis estadístico o visualizaciones.**


**Recorrer cada partido y calcular métricas básicas:**

```python
# Dataframe para guardar las métricas:
metricas_partidos = []

# Traer datos de sb.events:
for match_id in Mundial_2022['id_partido']:
    eventos = sb.events(match_id=match_id)

    # Equipos del partido:
    local = Mundial_2022.loc[Mundial_2022['id_partido']==match_id, 'equipo_local'].values[0]
    visitante = Mundial_2022.loc[Mundial_2022['id_partido']==match_id, 'equipo_visitante'].values[0]

    # Tiros:
    tiros_local = eventos[(eventos['team']==local) & (eventos['type']=='Shot')].shape[0]
    tiros_visitante = eventos[(eventos['team']==visitante) & (eventos['type']=='Shot')].shape[0]

    # Tiros al arco:
    tiros_al_arco_local = eventos[
        (eventos['team']==local) &
        (eventos['type']=='Shot') &
        (eventos['shot_outcome'].isin(['Saved', 'Goal']))
    ].shape[0]

    tiros_al_arco_visitante = eventos[
        (eventos['team']==visitante) &
        (eventos['type']=='Shot') &
        (eventos['shot_outcome'].isin(['Saved', 'Goal']))
    ].shape[0]

    # Pases:
    pases_local = eventos[(eventos['team']==local) & (eventos['type']=='Pass')].shape[0]
    pases_visitante = eventos[(eventos['team']==visitante) & (eventos['type']=='Pass')].shape[0]

    # Duelos ganados:
    duelos_ganados_local = eventos[
        (eventos['team']==local) &
        (eventos['type']=='Duel') &
        (eventos['duel_outcome']=='Won')
    ].shape[0]

    duelos_ganados_visitante = eventos[
        (eventos['team']==visitante) &
        (eventos['type']=='Duel') &
        (eventos['duel_outcome']=='Won')
    ].shape[0]

    # Faltas:
    faltas_local = eventos[(eventos['team']==local) & (eventos['type']=='Foul Committed')].shape[0]
    faltas_visitante = eventos[(eventos['team']==visitante) & (eventos['type']=='Foul Committed')].shape[0]

    # Guardar resultados:
    metricas_partidos.append({
        'id_partido': match_id,
        'tiros_local': tiros_local,
        'tiros_visitante': tiros_visitante,
        'tiros_al_arco_local': tiros_al_arco_local,
        'tiros_al_arco_visitante': tiros_al_arco_visitante,
        'pases_local': pases_local,
        'pases_visitante': pases_visitante,
        'duelos_ganados_local': duelos_ganados_local,
        'duelos_ganados_visitante': duelos_ganados_visitante,
        'faltas_local': faltas_local,
        'faltas_visitante': faltas_visitante
    })
```

**Crear DataFrame de métricas y unirlo al DataFrame principal:**

```python
df_metricas = pd.DataFrame(metricas_partidos)

# Unir al DataFrame principal:
Mundial_2022 = Mundial_2022.merge(df_metricas, on='id_partido', how='left')

# Ver resultado:
Mundial_2022.head()
```

## 1. ¿Cuánto más tiros al arco, mayor probabilidad de victoria?

  * Se compara qué equipo realizó más tiros al arco en cada partido con el resultado final, para ver si existe relación entre ambas variables.


```python
# Evitas advertencia (warning) de Seaborn:
import warnings
warnings.filterwarnings("ignore")

# Determinar qué equipo hizo más tiros al arco en cada partido:
def equipo_mas_tiros(row):
    if row['tiros_al_arco_local'] > row['tiros_al_arco_visitante']:
        return row['equipo_local']
    elif row['tiros_al_arco_visitante'] > row['tiros_al_arco_local']:
        return row['equipo_visitante']
    else:
        return 'Empate en tiros'

Mundial_2022['equipo_mas_tiros'] = Mundial_2022.apply(equipo_mas_tiros, axis=1)

# Verificar si el equipo con más tiros fue el ganador:
Mundial_2022['coincide_victoria_tiros'] = Mundial_2022.apply(
    lambda row: 'Coincide' if row['victoria'] == row['equipo_mas_tiros']
    else ('Empate' if row['victoria'] == 'Empate' or row['equipo_mas_tiros'] == 'Empate en tiros'
          else 'No coincide'),
    axis=1
)

# Calcular porcentaje de partidos:
resultado_tiros = Mundial_2022['coincide_victoria_tiros'].value_counts(normalize=True) * 100

# Gráfico:
plt.figure(figsize=(8,5))
sns.barplot(x=resultado_tiros.index, y=resultado_tiros.values, palette='coolwarm')
plt.title('¿Cuánto más tiros al arco, mayor probabilidad de victoria?', fontsize=14, weight='bold')
plt.xlabel('Relación entre tiros al arco y resultado del partido', fontsize=11)
plt.ylabel('Porcentaje de partidos (%)', fontsize=11)
plt.ylim(0, 100)

# Etiquetas sobre las barras:
for i, v in enumerate(resultado_tiros.values):
    plt.text(i, v + 1, f'{v:.1f}%', ha='center', fontweight='bold', fontsize=10)

plt.show()
```

**1. ¿Cuánto más tiros al arco, mayor probabilidad de victoria?**

  * En casi la mitad de los partidos (45.3%), el equipo que más tiros al arco realizó fue el que ganó.
Esto sugiere una tendencia positiva entre cantidad de tiros y probabilidad de victoria, aunque no es determinante, ya que en un 35.9% hubo empate y en un 18.8% ganó el rival con menos tiros.

## 2. ¿La cantidad de pases completados está relacionada con el resultado?

  * En este caso comparamos si el equipo que completó más pases en cada partido fue también el que obtuvo la victoria.

```python
# Evitas advertencia (warning) de Seaborn:
import warnings
warnings.filterwarnings("ignore")

# Determinar si el equipo con más pases fue el ganador:
def comparar_pases(row):
    if row['pases_local'] > row['pases_visitante']:
        equipo_mas_pases = row['equipo_local']
    elif row['pases_local'] < row['pases_visitante']:
        equipo_mas_pases = row['equipo_visitante']
    else:
        return "Empate"

    if row['victoria'] == "Empate":
        return "Empate"
    elif equipo_mas_pases == row['victoria']:
        return "Coincide"
    else:
        return "No coincide"

# Aplicar la función:
Mundial_2022["relacion_pases_victoria"] = Mundial_2022.apply(comparar_pases, axis=1)

# Calcular proporciones:
resultado_pases = Mundial_2022["relacion_pases_victoria"].value_counts(normalize=True) * 100

# Gráfico:
plt.figure(figsize=(7,5))
sns.barplot(x=resultado_pases.index, y=resultado_pases.values, palette="coolwarm", legend=False)

plt.title("¿Más pases completados = Mayor probabilidad de victoria?", fontsize=13, weight='bold')
plt.ylabel("Porcentaje de partidos (%)")
plt.xlabel("Relación entre pases completados y resultado del partido")

# Etiquetas con porcentajes:
for i, valor in enumerate(resultado_pases.values):
    plt.text(i, valor + 1, f"{valor:.1f}%", ha='center', fontsize=11, fontweight='bold')

plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

**2. ¿La cantidad de pases completados está relacionada con el resultado?**

  * La cantidad de pases completados no tiene una relación clara y fuerte con el resultado del partido. Es decir, más pases completados no necesariamente implica mayor probabilidad de victoria.

## 3. ¿Qué impacto tienen las faltas en el desarrollo del partido?

  * Se calculó qué equipo cometió más faltas en cada partido y se comparó con el equipo que perdió. Luego se obtuvo el porcentaje de partidos donde el equipo con más faltas coincidió con la derrota, no coincidió o hubo empate.

```python
# Evitas advertencia (warning) de Seaborn:
import warnings
warnings.filterwarnings("ignore")

# Determinar qué equipo hizo más faltas en cada partido:
def equipo_mas_faltas(row):
    if row['faltas_local'] > row['faltas_visitante']:
        return row['equipo_local']
    elif row['faltas_visitante'] > row['faltas_local']:
        return row['equipo_visitante']
    else:
        return 'Empate en faltas'

Mundial_2022['equipo_mas_faltas'] = Mundial_2022.apply(equipo_mas_faltas, axis=1)

# Verificar si el equipo con más faltas fue el que perdió:
Mundial_2022['coincide_derrota_faltas'] = Mundial_2022.apply(
    lambda row: 'Coincide' if row['derrota'] == row['equipo_mas_faltas']
    else ('Empate' if row['derrota'] == 'Empate' or row['equipo_mas_faltas'] == 'Empate en faltas'
          else 'No coincide'),
    axis=1
)

# Calcular porcentaje de partidos:
resultado_faltas = Mundial_2022['coincide_derrota_faltas'].value_counts(normalize=True) * 100

# Gráfico de barra horizontal:
plt.figure(figsize=(8,5))
sns.barplot(x=resultado_faltas.values, y=resultado_faltas.index, palette='coolwarm')
plt.title('¿Más faltas = mayor probabilidad de perder?', fontsize=14, weight='bold')
plt.xlabel('Porcentaje de partidos (%)', fontsize=11)
plt.ylabel('Relación entre faltas y derrota', fontsize=11)
plt.xlim(0, 100)

# Etiquetas sobre las barras:
for i, v in enumerate(resultado_faltas.values):
    plt.text(v + 1, i, f'{v:.1f}%', va='center', fontweight='bold', fontsize=10)

plt.show()
```

3. ¿Qué impacto tienen las faltas en el desarrollo del partido?

  * Las faltas tienen un impacto moderado en el desarrollo del partido: cometer más faltas no garantiza la derrota, pero sí se observa una ligera tendencia a que el equipo que falta más tenga más riesgo de perder. En otras palabras, las faltas pueden influir, pero no son un factor determinante por sí solas.

## 4. ¿Ganar los duelos puede darte una victoria?

  * Se compara partido por partido, qué equipo ganó más duelos y si eso coincidió con el ganador.

```python
# Evitas advertencia (warning) de Seaborn:
import warnings
warnings.filterwarnings("ignore")

# Determinar qué equipo ganó más duelos en cada partido:
def equipo_mas_duelos(row):
    if row['duelos_ganados_local'] > row['duelos_ganados_visitante']:
        return row['equipo_local']
    elif row['duelos_ganados_visitante'] > row['duelos_ganados_local']:
        return row['equipo_visitante']
    else:
        return 'Empate en duelos'

Mundial_2022['equipo_mas_duelos'] = Mundial_2022.apply(equipo_mas_duelos, axis=1)

# Verificar si el equipo con más duelos fue el ganador:
Mundial_2022['coincide_victoria_duelos'] = Mundial_2022.apply(
    lambda row: 'Coincide' if row['victoria'] == row['equipo_mas_duelos']
                else ('Empate' if row['victoria'] == 'Empate' or row['equipo_mas_duelos'] == 'Empate en duelos'
                      else 'No coincide'), axis=1
)

# Calcular porcentaje de partidos:
resultado_duelos = Mundial_2022['coincide_victoria_duelos'].value_counts(normalize=True) * 100

# Gráfico:
plt.figure(figsize=(8,5))
sns.barplot(x=resultado_duelos.values, y=resultado_duelos.index, palette='coolwarm')
plt.title('¿Ganar los duelos puede darte una victoria?', fontsize=14, weight='bold')
plt.xlabel('Porcentaje de partidos (%)', fontsize=11)
plt.ylabel('Relación entre duelos ganados y resultado', fontsize=11)
plt.xlim(0, 100)

# Etiquetas sobre las barras:
for i, v in enumerate(resultado_duelos.values):
    plt.text(v + 1, i, f'{v:.1f}%', va='center', fontweight='bold', fontsize=10)

plt.show()
```

4. ¿Ganar los duelos puede darte una victoria?

  * Ganar los duelos aumenta algo las chances de victoria, pero no garantiza el triunfo, En otras palabras, es importante, pero no determinante.

## Conclusión general:

  * Los factores como tiros al arco, duelos ganados o faltas tienen cierta influencia en el resultado de un partido, pero ninguno por sí solo determina la victoria. La cantidad de pases completados, en particular, no muestra relación clara con ganar o perder. En conjunto, esto indica que el fútbol es un deporte multifactorial: las estadísticas individuales reflejan tendencias, pero la victoria depende de la combinación de varias variables y del contexto del partido, no de un solo indicador.
